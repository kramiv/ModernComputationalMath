# Yahtzee in Python

## Rules of Yahtzee (simplified)

Roll five dice, and re-roll some of them until all five dice show the same number.

1. __Roll five dice.__ Suppose we get 1, 2, 1, 3, 1.
2. __Keep the values that occur the most.__ In our example, we have three 1s. 
3. __Re-roll the other dice.__ In this case, we re-roll the two dice that were not 1s. Suppose we get 1 and 4.
4. __Repeat steps 2 and 3 until all dice show the same number.__ In our example, we now have 1,1,1,1,4. Re-roll the die that isn't a 1 until it shows a 1

We want to keep track of how many times we roll the dice until all five dice show the same number.

We need to simulate rolling a die. We will use the **randint** function, which is found in the **random** module.

```
import random
random.randint(1,6)
```

Now we will write a function to roll $n$ dice.

```
def rollDice(n):
  rolls = [random.randint(1,6) for i in range (n)]
  return rolls
```

Note that the running the code above doesn't produce any output, it just stores our function in memory. Now we can try out our function to roll 5 dice:


```
currRolls = rollDice(5)
print(currRolls)
```

```
counts = [currRolls.count(i) for i in range (1,7)]#6 is 0 1 2 3 4 5
print(counts)
```
